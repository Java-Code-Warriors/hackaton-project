```python
import sqlite3
import pandas as pd
import json

def connect_to_db(db_name):
    """Establishes a connection to the SQLite database.

    Args:
        db_name: The path to the SQLite database file.

    Returns:
        A database connection object, or None if an error occurs.
    """
    try:
        conn = sqlite3.connect(db_name)
        return conn
    except sqlite3.Error as e:
        print(f"Error connecting to database: {e}")
        return None

def get_products(conn, gender, budget, goal):
    """Retrieves product recommendations from the database.

    Args:
        conn: The database connection object.
        gender: The gender ('М' or 'Ж').
        budget: The budget category (string).
        goal: The gift purpose (string).

    Returns:
        A Pandas DataFrame containing the product recommendations, or an empty DataFrame if no products are found or an error occurs.
    """
    if conn is None:
        return pd.DataFrame()

    table = 'Man' if gender == 'М' else 'Women' if gender == 'Ж' else None
    if table is None:
        return pd.DataFrame()

    query = f"SELECT Name, ProductLink FROM {table} WHERE Budget = ? AND Goal = ?"
    try:
        df = pd.read_sql_query(query, conn, params=(budget, goal))
        return df
    except pd.io.sql.DatabaseError as e:
        print(f"Database error: {e}")
        return pd.DataFrame()
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return pd.DataFrame()


def validate_input(gender, budget, goal):
    """Validates user input.

    Args:
        gender: The gender.
        budget: The budget.
        goal: The gift purpose.

    Returns:
        A dictionary containing error messages if input is invalid, or None if input is valid.
    """
    errors = {}
    if gender not in ['М', 'Ж']:
        errors["gender"] = "Invalid gender input."
    if budget not in ["До 2000", "До 7000", "До 15000", "Более 15000"]:
        errors["budget"] = "Invalid budget input."
    if goal not in ["Гаджеты", "Для дома", "Для учебы/офиса", "Игры и аксессуары"]:
        errors["goal"] = "Invalid gift purpose input."
    return errors


def proccesing_ml(gender, budget, goal):
    """Main function to handle product recommendations."""
    db_name = 'database.db'  # Database file name
    conn = connect_to_db(db_name)
    if conn is None:
        return

    errors = validate_input(gender, budget, goal)
    if errors:
        print(json.dumps({"errors": errors}))
        conn.close()
        return

    products_df = get_products(conn, gender, budget, goal)
    products_list = [{"name": row['Name'], "link": row['ProductLink']} for _, row in products_df.iterrows()]

    print(json.dumps({"products": products_list}, ensure_ascii=False, indent=4))
    conn.close()


if __name__ == "__main__":
    gender = input("Какой у вас пол? (М/Ж): ")
    budget = input("Какой у вас бюджет? (До 2000, До 7000, До 15000, Более 15000): ")
    goal = input("Для чего подарок? (Гаджеты, Для дома, Для учебы/офиса, Игры и аксессуары): ")
    main(gender, budget, goal)
```
